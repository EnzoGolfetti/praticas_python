```python
df_pessoa = venda_veic[['cod_pessoa', 'nome', 'genero', 'data_nascimento']].drop_duplicates()
```

```python
df_modelo = venda_veic[['modelo', 'fabricante']].drop_duplicates()
fabricantes = list(venda_veic.fabricante.unique())
```

```python
for pessoa in df_pessoa.itertuples(): #aqui unimos em tuplas os dados do df_pessoa para ficar mais fácil na hora do query
    print(pessoa)
```

```python
#aqui é a hora que inserimos os dados na tabela lá no SQl
#pessoa é a tabela que nomeamos no MySQL
#atribuimos os dados das nossas tuples às colunas que criamos no MySQl através do 'try' e do 'except' 

for pessoa in df_pessoa.itertuples(): 
    cod_pessoa = pessoa.cod_pessoa
    nome = pessoa.nome
    genero = pessoa.genero
    data_nascimento = str(pessoa.data_nascimento)
    try:
        query = f""" insert into pessoa (cod_pessoa, nome, genero, data_nascimento)
                     values ('{cod_pessoa}', '{nome}', '{genero}', '{data_nascimento}')"""
        conn.execute(query)
        print(f'Registro inserido com sucesso: {nome}')
    except Exception as e:
            print(f"Não foi possível inserir o registro de: {nome}, o erro encontrado foi: {e}")
```

```python
for fabricante in fabricantes:
    try:
        query = f"""insert into fabricante (fabricante)
                    values ('{fabricante}')"""
        conn.execute(query)
        print(f'Registro inserido com sucesso: {fabricante}')
    except Exception as e:
        print(f'Não foi possível inserir o registro: {fabricante}. O erro foi: {e}')
```

```python
df_fabricante = pd.read_sql('fabricante', con=conn) #essa função lê os dados no banco de dados MySQL
```
