Quais foram os produtos analisados nessa coleta?

```python
hs_code = dataset['Commodity Code'].value_counts()
print('O HS Code dos produtos analisados são:')
print('inclui a quantidade de exportações registradas por cada um de um total de 662')
hs_code.to_frame()
```

Qual o nome de cada uma dessas classificações?

```python
nome_commodity = dataset['Commodity'].value_counts()
print('Os produtos analisados foram:')
nome_commodity.to_frame()
```

Qual país mais recebeu exportações do Brasil no período analisado?

```python
group_pais = dataset.groupby('Partner')['Qty'].sum().sort_values(ascending=False)
```

```python
group_pais = pd.DataFrame(group_pais).reset_index()
#atenção: se não resetar o index não funcionar passar pra Dataframe nem futuras observações
```

```python
group_pais.head(11)
```

```python
print(f'O país que mais importou no período foi: {group_pais.iloc[1,0]}')
```

Quais os 10 países que mais importaram açúcar do Brasil entre 2016 e 2019?

```python
dez_mais = group_pais.iloc[1:11,0]
dez_mais.to_frame()
```

Qual o total de exportações de açúcar para os Estados Unidos da América (USA) no período?

```python
total_eua = dataset[dataset['Partner ISO'] == 'USA']['Qty'].sum()
print(f'O total de exportações de açúcar para os EUA no período 2016-2019 foi: {total_eua}')
```

Qual o total de exportações para a China (CHN) no período?

```python
total_chn = dataset[dataset['Partner ISO'] == 'CHN']['Qty'].sum()
print(f'O total de exportações de açúcar para a China no período 2016-2019 foi: {total_chn}')
```

Qual a média de exportação do Brasil no perídodo?

```python
media_expo = dataset[dataset['Reporter ISO'] == 'BRA']['Qty'].mean()
print(f'A média de exportações no período foi: {media_expo:.0f}')
```

Qual o preço médio das exportações?

```python
dataset['Trade Value (US$)'].isnull().sum()
```

```python
media_trade = dataset['Trade Value (US$)'].mean()
print(f'A média do preço no período foi de: {media_trade:.0f}')
```

Com o frame acima podemos ter uma conclusão inicial de que as exportações vinham caindo.

Porém agora, vamos analisar se houve aumento nos preços ao longo dos anos o que pode ter justificado a diminuião das exportações mas com continuidade dos ganhos para o produtor brasileiro.

```python
ganhos_year = dataset.groupby('Year')['Trade Value (US$)'].sum().sort_values(ascending=False)
ganhos_year = ganhos_year.to_frame().reset_index()
```

```python
ganhos_year
```

Podemos observar que fora um aumento no valor das exportações em 2017, no resto do período a queda nos valores acompanhou a queda na quantidade, indicando efetiva dimininuição das exportações e dos ganhos dos produtores brasileiros. 

Mas observar o Trade Value geralmente não é boa medida, seria melhor se tivessemos os valores FOB!
