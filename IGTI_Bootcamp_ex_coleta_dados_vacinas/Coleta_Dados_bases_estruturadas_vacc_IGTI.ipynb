# Pergunta 3: Qual é a vacina mais utilizada no mundo?

```python
vacinas = df_vac.groupby('vacinas')
```

```python
vacinas_mais_usadas = vacinas['pais'].nunique().sort_values(ascending=False)
```

```python
vacinas_mais_usadas #posso filtrar para ver as mais usadas
```

```python
df_vacinas_mais = vacinas_mais_usadas.to_frame().reset_index() #o reset_index põe um index no dataframe
```

```python
df_vacinas_mais.rename(columns={'pais': 'Qtd. países usam'}, inplace=True)
```

```python
df_vacinas_mais
```

```python
pais_vacina = paises['vacinas'].unique()
pais_vacina = pais_vacina.to_frame().reset_index()
pais_vacina
```

# Pergunta 4: Qual país possui o maior número de vacinados por dia?

```python
df_vac.head()
```

```python
paises = df_vac.groupby('pais')
```

```python
paises_vac_diaria_max = paises['vacinaçoes_diarias'].max().sort_values(ascending=False)
```

# Pergunta 5: Qual a média de vacinação por dia por país?

```python
media_vac_diaria = paises['vacinaçoes_diarias'].mean().sort_values(ascending=False)
```

```python
media_vac_diaria = media_vac_diaria .to_frame().reset_index()
media_vac_diaria 
```
